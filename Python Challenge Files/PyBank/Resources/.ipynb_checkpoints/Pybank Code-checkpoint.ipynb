In this Challenge, you are tasked with creating a Python script to analyze the financial records of your company. You will be given a financial dataset called budget_data.csv. The dataset is composed of two columns: "Date" and "Profit/Losses".

Your task is to create a Python script that analyzes the records to calculate each of the following values:

The total number of months included in the dataset

The net total amount of "Profit/Losses" over the entire period

The changes in "Profit/Losses" over the entire period, and then the average of those changes

The greatest increase in profits (date and amount) over the entire period

The greatest decrease in profits (date and amount) over the entire period

Your analysis should align with the following results:

```python
#Import 
import os
import csv
```

```python
#create a path to the file 
budget_data = os.path.join('../Resources/budget_data.csv')
```

```python
#create variables and print the data 
month = 0
prof_loss_total = 0
tot_change = 0
max_change = 0
min_change = 0

with open(budget_data) as budget_data_file:
    #csv reader tells specific variables and delimiter that holds content
    budget_reader = csv.reader(budget_data_file, delimiter=',')
    
    print(budget_reader)
    #Read the header row fist but skip if file does not have header
    csv_header = next(budget_reader)
    print(f"Budget Data:{csv_header}")
    first_month = next(budget_reader)
    month = month + 1
    prev_pl = int(first_month[1])
    prof_loss_total = prof_loss_total + prev_pl
    
    #find the total months and the changes 
    for rows in budget_reader:
        print(rows)
        month = month + 1
        prof_loss = int(rows[1])
        prof_loss_total = prof_loss_total + prof_loss
        change = prof_loss - prev_pl
        prev_pl = prof_loss
        tot_change = tot_change + change
        if change > max_change:
            max_change =change
            first_month = rows[0]
        if change < min_change:
            min_change = change
            d_month = rows[0]
        avg_change = round((tot_change/(month -1)),2)
        #print out the items we are looking for 
        print('-------------------------------------')
        print('Financial Analysis')
        print(f'Total month:{month}')
        print(f'Profit/Loss Total:{prof_loss_total}')
        print(f'Average Change:{avg_change}')
        print(f'Greatest Increase in Profits:{first_month}{max_change}')
        print(f'Greatest Decrease in Profits:{d_month}{min_change}')
                        
```

```python
output_budget=os.path.join("..","Resources","pybank_output.csv")
```

```python
with open(output_budget,'w')as budget_data_file:
    #creating a csv file and csv.writer
    budget_writer=csv.writer(budget_data_file)
    #write out the results in csv file
    budget_writer.writerow(['Financial Analysis: '])
    budget_writer.writerow([f'Total month:{month}'])
    budget_writer.writerow([f'Profit/Loss Total: ${prof_loss_total}'])
    budget_writer.writerow([f'Average Change: ${avg_change}'])
    budget_writer.writerow([f'Greatest Increase in Profits:{month} ${max_change}'])
    budget_writer.writerow([f'Greatest Decrease in Profits: {d_month} ${min_change}'])
```
